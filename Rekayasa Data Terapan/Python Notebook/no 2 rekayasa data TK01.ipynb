```python
all_docs = []
testfile1 = open("test.txt", "r")
testfile2 = open("test2.txt", "r")
testfile3 = open("test3.txt", "r")
testfile4 = open("test4.txt", "r")
all_docs.append(testfile1.read())
all_docs.append(testfile2.read())
all_docs.append(testfile3.read())
all_docs.append(testfile4.read())
all_docs
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

tfidf_vectorizer = TfidfVectorizer()
values = tfidf_vectorizer.fit_transform(all_docs)

# Show the Model as a pandas DataFrame
feature_names = tfidf_vectorizer.get_feature_names()
pd.DataFrame(values.toarray(), columns = feature_names)
```
